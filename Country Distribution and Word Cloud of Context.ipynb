```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
file_path = r'C:\Users\hp\Documents\Collecting Data Lab\week4\times-person-of-the-year.csv'
times_person_of_the_year_df = pd.read_csv(file_path)
```

```python
# What is the country distribution of the awards?
```

```python
import matplotlib.pyplot as plt

country_counts = times_person_of_the_year_df['Country'].value_counts()

plt.figure(figsize=(12, 8))
country_counts.plot(kind='bar')
plt.title('Distribution of Time Person of the Year by Country')
plt.xlabel('Country')
plt.ylabel('Number of Times Chosen')
plt.xticks(rotation=90)  
plt.tight_layout() 
plt.show()
```

```python
country_counts = times_person_of_the_year_df['Country'].value_counts()

plt.figure(figsize=(10, 10)) 
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Time Person of the Year Distribution by Country')
plt.axis('equal') 
plt.show()
```

```python
# Wow! Most of the yearly person comes from USA
```

```python
#Then, here is the word cloud for 'Context'
```

```python
from wordcloud import WordCloud

text = " ".join(context for context in times_person_of_the_year_df['Context'].dropna())

wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)

plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud of 'Context' in Time Person of the Year")
plt.show()
```
